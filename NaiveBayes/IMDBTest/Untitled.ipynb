```python
import pandas as pd
```

```python
movies_train = pd.read_csv("movie_review_train.csv")
```

```python
movies_train.shape
```

```python
movies_train.head()
```

```python
movies_train['class'].value_counts()
```

```python
movies_train['rev'] = movies_train['class'].map({"Pos":1, "Neg":0})
```

```python
movies_train.head()
```

```python
movies_train = movies_train.drop(["class"],axis=1)
```

```python
movies_train
```

```python
x_train = movies_train.text
y_train = movies_train.rev 
```

```python
#from sklearn.model_selection import train_test_split
# split not needed as we already have train test split
```

```python
#Feature Extraction
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer( stop_words='english')
```

```python
vect.fit(x_train)
```

```python
vect.vocabulary_
```

```python
len(vect.vocabulary_)
```

```python
len(vect.get_feature_names())
```

```python
vect = CountVectorizer( stop_words='english', min_df=.03, max_df=.8)
```

```python
vect.fit(x_train)
```

```python
len(vect.get_feature_names())
```

```python
test_data = pd.read_csv("movie_review_test.csv")
```

```python
test_data['rev'] = test_data['class'].map({"Pos":1, "Neg":0})
```

```python
test_data = test_data.drop(['class'], axis=1)
```

```python
x_test = test_data.text
y_test = test_data.rev
```

```python
x_train_transformed = vect.transform(x_train)
x_test_transformed = vect.transform(x_test)
```

```python
x_test_transformed
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
print(x_test_transformed.shape)
```

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train_transformed, y_train)
```

```python
y_pred_rev = bnb.predict(x_test_transformed)
```

```python
y_pred_proba = bnb.predict_proba(x_test_transformed)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_pred_rev)
```

```python
metrics.confusion_matrix(y_test, y_pred_rev)
```
